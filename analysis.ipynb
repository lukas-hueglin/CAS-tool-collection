# Analysis

```python
import spb
import sympy
import matplotlib.pyplot as plt
```

## Define symbols and functions

```python
# define Symbols
x = sympy.Symbol('x')
y = sympy.Symbol('y')

#define Functions
f = sympy.Lambda(y, x**2)
```

## Differentiate

```python
# differentiate
df = sympy.Lambda(y, sympy.diff(f(x), x))

# output
df
```

## Integrate

### Indefinite Integral

```python
# integrate
F = sympy.Lambda(x, sympy.integrate(f(x), x))

# output
F
```

### Definite Integral

```python
# define Symbols for limits
a = sympy.Symbol('a')
b = sympy.Symbol('b')

# integrate with limits
A = sympy.Lambda((a, b), sympy.integrate(f(x), (x, a, b)))

# output
A
```

## Plotting

```python
# set the function to plot
FUNCTIONS = [f(x), df(x), F(x)]

# set the plot style to dark_background
plt.style.use(plt.style.available[6])

# create empty plot
p = spb.plot(use_latex=True, show=False)

# add all functions from FUNCTIONS
for func in FUNCTIONS:
    p.extend(spb.plot(func, show=False))

# set left and lower axis to zero
p.ax.spines['left'].set_position('zero')
p.ax.spines['bottom'].set_position('zero')

# eliminate upper and right axes
p.ax.spines['right'].set_color('none')
p.ax.spines['top'].set_color('none')

# show ticks in the left and lower axes only
p.ax.xaxis.set_ticks_position('bottom')
p.ax.yaxis.set_ticks_position('left')

# set label position to right and top
p.ax.set_xlabel(p.ax.get_xlabel(), loc='right')
p.ax.set_ylabel(p.ax.get_ylabel(), loc='top')

# set position of the legend
p.ax.legend(loc = 'upper right')

# remove grid
p.ax.grid(False)

# set dpi of plot
p.fig.set_dpi(120)
```
